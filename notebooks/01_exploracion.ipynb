```python
import pandas as pd

# Definimos la ruta relativa desde la carpeta 'notebooks' hasta 'data/raw'
ruta_archivo = '../data/raw/ecommerce_sales_raw.csv'

# Cargamos los datos
df = pd.read_csv(ruta_archivo)

# Mostramos las primeras 5 filas para comprobar que ha cargado bien
display(df.head())

# Vemos la estructura, tipos de datos y valores nulos
print("\n--- Información del Dataset ---")
df.info()

print("\n--- Conteo de Valores Nulos ---")
print(df.isnull().sum())
```

```python
# Hacemos una copia para trabajar sobre seguro
df_cleaned = df.copy()

# 1. Convertir la columna de fecha a un formato que Python entienda como "Fecha"
df_cleaned['order_date'] = pd.to_datetime(df_cleaned['order_date'])

# 2. Extraer partes de la fecha para nuestros futuros gráficos
df_cleaned['year'] = df_cleaned['order_date'].dt.year
df_cleaned['month_num'] = df_cleaned['order_date'].dt.month # Para ordenar cronológicamente
df_cleaned['month_name'] = df_cleaned['order_date'].dt.strftime('%B') # Para mostrar "January", "February"...
df_cleaned['day_of_week'] = df_cleaned['order_date'].dt.strftime('%A') # Para saber si compran más en finde o entre semana

# 3. Eliminar filas duplicadas por si acaso existen
antes_duplicados = df_cleaned.shape[0]
df_cleaned = df_cleaned.drop_duplicates()
print(f"Filas eliminadas por duplicados: {antes_duplicados - df_cleaned.shape[0]}")

# Verificamos cómo quedan nuestras nuevas columnas
df_cleaned[['order_date', 'year', 'month_name', 'day_of_week']].head()
```
